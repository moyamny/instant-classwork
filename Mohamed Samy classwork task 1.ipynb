## WAP to enter a number from user and print its absolute value.*

```python
#####first way using abs function

negativeNum = input("please enter your number : ")
while not negativeNum.isdigit():
    negativeNum = input("please enter a number: ")
else:
    negativeNum = int(negativeNum)
    if negativeNum > 0:
        print("your number is absolute")
    elif negativeNum == 0:
        print("zero is absolute")
    else:
        print(abs(negativeNum))
```

```python
#####second way by arthimtic math symbol (-)
negativeNum = input("please enter your number : ")
while not negativeNum.isdigit():
    negativeNum = input("please enter a number: ")
else:
    negativeNum = int(negativeNum)
    if negativeNum > 0:
        print("your number is absolute")
    elif negativeNum == 0:
        print("zero is absolute")
    else:
        print(-negativeNum)
```

##  WAP to enter a year from user and check if it is leap year of not.

```python
year = input("please enter your year: ")

while not year.isdigit():
    year = input("please enter a number: ")
else:
    year = int(year)
    if year % 4 == 0:
        print(year ,"leap year")
    else:
        print(year ,"not a leap year")

```

## WAP to enter the age of 3 person and print oldest and yougest among them.

```python
while True:
    try:
        age1 = int(input("Enter age of person 1: "))
        age2 = int(input("Enter age of person 2: "))
        age3 = int(input("Enter age of person 3: "))

        if age1 < 0 or age2 < 0 or age3 < 0:
            print("Age cannot be negative. Please enter valid ages.")
            continue
        break
    except ValueError:
        print("Age must be an integer number. Please enter valid ages.")

print("The oldest age is:", max(age1, age2, age3))
```

## Write a program that asks the user to enter a series of numbers. The program should ask for numbers until the user enters a negative number. Once the user enters a negative number, the program should stop and print the sum of all the positive numbers entered

## Example


### Enter a number: 5
### Enter a number: 8
### Enter a number: 12
### Enter a number: -2
### The sum of the positive numbers is 25.

```python
pNumSum = 0

while True:
    try:
        num = int(input("Please enter a number: "))

        if num > 0:
            pNumSum += num
            continue
        break

    except ValueError:
        print("Invalid input. Please enter a valid number.")

print("The sum of the positive numbers you entered is:", pNumSum)
```

## Write a program that asks the user to enter their name and age. The program should check if the user's age is greater than or equal to 18. If it is, the program should print a message saying they are eligible to vote. Otherwise, the program should print a message saying they are not eligible to vote due to being underage.


```python
name = input("enter your name : ")

while True:
    try:
       age = float(input("enter your age : "))

       if age >= 18:
           print(name,"you are eligible to vote")
           break
       else:
           print(name,"you are not eligible to vote due to being underage")
           break
    except ValueError:
        print("invalid age input ,Please enter a number")
```
