**7-Write a Python program to count the occurrences of each word in a given sentence (Sample String:amr and ahmed are frindes but amr is the tallest Expected Result:2)

```python
def countwords(sentence):
    words = sentence.split()
    counts = {}
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts

countwords("ahmed and mohamed are friends but ahmed is the tallest")
```

**12- Write a Python program to add prefix text to all of the lines in a string

```python
def addprefix(string, prefix):
    lines = string.split("\n")
    new_lines = []
    for line in lines:
        new_lines.append(prefix + line)
    return "\n".join(new_lines)

addprefix("ahmed\nmohamed", "Mr.")
```

 **17-Write a Python program to count repeated characters in a string (hint:use dictionary)

```python
def countrepeatedchars(string):
    counts = {}
    for char in string:
        if char in counts:
            counts[char] += 1
        else:
            counts[char] = 1

    repeated = {}
    for char, count in counts.items():
        if count > 1:
            repeated[char] = count
    return repeated

countrepeatedchars("aloo ya abo el so7ab")
```

**18-Write a Python program to find the first non-repeating character in a given string

```python
def nonrepeatingchar(string):
    counts = {}
    for char in string:
        if char in counts:
            counts[char] += 1
        else:
            counts[char] = 1

    for char in string:
        if counts[char] == 1:
            return char

nonrepeatingchar("allooo")
```

Output : [4, 3] )

```python
def extractfreq(lst, k):
    counts = {}
    for elem in lst:
        if elem in counts:
            counts[elem] += 1
        else:
            counts[elem] = 1
            result = []
    for elem, count in counts.items():
        if count > k:
            result.append(elem)
    return result

extractfreq([9,9,9,9,8,8,8,7,7,7],3)
```

**32-write a python program to find the Strongest Neighbour (Input: 1 2 2 3 4 5
Output: 2 2 3 4 5)

```python
def findstrongestneighbour(lst):
    result = []
    for i in range(1, len(lst)):
        if lst[i] > lst[i-1]:
            result.append(lst[i])
        else:
            result.append(lst[i-1])
    return result

findstrongestneighbour([1, 2, 2, 3, 4, 5])
```

Output:
1 2 3 ##
1 3 2 ##
2 1 3 ##

```python
def print_combinations(digits):

  for i in range(3):
    for j in range(3):
      if j == i:
        continue
      for k in range(3):
        if k == i or k == j:
          continue
        print(digits[i], digits[j], digits[k])

print_combinations([1, 2, 3])
```

**35-write a Python program to get all unique combinations of two Lists (List_1 = ["a","b"]
List_2 = [1,2]
Unique_combination = [[('a',1),('b',2)],[('a',2),('b',1)]] )

```python
def get_unique_combinations(list1, list2):
    unique_combinations = []
    for item1 in list1:
        for item2 in list2:
            combination = (item1, item2)
            if combination not in unique_combinations:
                unique_combinations.append(combination)
    return unique_combinations

get_unique_combinations(["a","b"],[1,2])
```

Output : [(4, 5, 5, 4)]
Input : test_list = [(4, 5, 5, 4), (5, 4, 3)], K = 5, N = 3
Output : [] )

```python
def retain_records(lst, k, n):
    result = []
    for elem in lst:
        if elem.count(k) == n:
            result.append(elem)
    return result

retain_records([(4, 5, 5, 4), (5, 4, 3)], 4, 1)
```

**46- Write a Python program to drop empty items from a given dictionary.
Original Dictionary:
{'c1': 'Red', 'c2': 'Green', 'c3': None}
New Dictionary after dropping empty items:
{'c1': 'Red', 'c2': 'Green'}

```python
def dropempty(dic):
    new_dic = {}
    for key, value in dic.items():
        if value is not None:
            new_dic[key] = value
    return new_dic

dropempty({'c1': 'Red', 'c2': 'Green', 'c3': None})
```
